# Distribución <font color='red'>Chi-cuadrado</font>

1. siempre es >=0. Por ej., nunca se puede calcular: (x2 <= -1)
2. deriva de la distribución normal, para muestras pequeñas (<= 30)
3. es la suma de las diferencias cuadrádicas relativas entre los valores reales y los predichos
4. el parámetro `k` representa los grados de libertad (df = dregrees of freedom)
5. demostración: ¿son independientes las variables?

**cuando el valor de n (grados de libertad) es mayor, más se asemeja a una distribución normal**


Se debe efectuar un pre análisis de la data, considerando:

1. MAYOR O IGUAL: Si **P(<font color='green'>x2(3)</font> >= 2.16)** [si la probabilidad de que x2 con 3 grados de libertad es mayor o igual a 2.16], ir a la tabla y buscar el resultado. En este caso, es 0.975 (97.5% de probabilidad). 
2. MENOR O IGUAL: Si **P(<font color='green'>x2(15)</font> < 11.04):** si la probabilidad buscada debe ser < (menor), no podemos acudir a la tabla. Se aplica la lógica inversa, vista en las otras distribuciones: **P(x2(15) < 11.04) = <font color='red'>1 - </font>P(x2(15) >= 11.04)**. En este caso = <font color='red'>1 -</font> 0.75 = 0.25 = 25%
3. ENTRE DOS VALORES: Si **P(11.59 < <font color='green'> x2(21) </font> <= 20.34)** [probabilidad de que x2 (con 21 grados de libertad) sea mayor a 11.59 y menor o igual a 20.34]
se realizan dos operaciones:

    a) **P(<font color='green'>x2(21)</font> >= 11.59)**: como en este caso se busca mayor o igual, recurrimos a la tabla: 0.95 (95%)
    
    b) **P(<font color='green'>x2(21)</font> >= 20.34)**: se invierte el <= por >= y vamos a la tabla: 0.5 (50%)
    
    c) se efectua una resta: **P(<font color='green'>x2(21)</font> >= 11.59) - P(<font color='green'>x2(21)</font> >= 20.34)** = 0.95 - 0.5 = 0.45 (45%)


## ejemplo:
tenemos dos variables:
1. Sencación de bienestar: 2 posibles valores: si/no
2. Práctica deportiva: 2 posibles valores: si/no
3. Requerimos un nivel de significancia = 0.01 (esto es el grado de confianza para rechazar la hipótesis nula)

¿se encuentran relacionadas ambas variables?


## prueba de hipótesis:
* H0 (hipótesis nula) = las variables son independientes
* H1 (alternativa) = las variables NO son independientes
