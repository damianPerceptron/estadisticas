# 1. <font color='green'>distribución uniforme</font>
* todos los valores tienen la misma probabilidad de ocurrir
* X tiene una distribución uniforme sobre el intervalo real (a, b), con a < b, si su función de densidad es:
    + f(x) = {1/b-a} --> a < x < b, en cualquier otro caso, la f(x)=0
* **su media (esperanza) es: (b + a) / 2**
* **su varianza es: (b-a)^2 / 12**

* sea X una v.a. U (intervalo) va a estar entre -1 y 1. 
    * a: va a ser -1; 
    * b: va a ser 1, por lo tanto el parámetro loc=-1 y scale=b-a = 1 - (-1) = 2

## 1.1 ejercicios:

## 1.1.1
Let's say we have a shoe store that sells pairs of shoes with a uniform probability of a sale throughout a seven-day week. That is, it's equally probable that a pair of shoes is sold on Monday as on Tuesday. The minimum number of pairs sold per week is 50, and the max is 250. Calculuate the probability of selling between 100 and 150 pairs of shoes. Calculate the mean and standard deviation

$$P_{100-150} = (150 - 50) \cdot \frac{1}{250 - 50} = \frac{100}{200} = \frac{1}{2}$$

$$\bar{x} = \frac{a + b}{2} = \frac{50 + 250}{2} = 150 \; \text{pairs}$$

$$ \begin{align} \sigma^2 &= \frac{(b - a)^2}{12} \\ \\ &= \frac{(250 - 50)^2}{12} = 3,333 \\ \\ &\text{so }\; \sigma = \sqrt{3333} = 58. \end{align}$$

<font color='blue'>resolución</font>
* la pdf con scipy
* mean, var  std manual

```python
from scipy.stats import uniform
x = 150-50 #cargo en x el rango de ventas
a = 50 #ventas mínimas x semana
b = 250 #ventas máximas x semana
print("pdf: ",uniform.pdf(x, loc=a, scale=(b-a))*100)
print("cdf: ",uniform.cdf(x, loc=a, scale=(b-a))*100)
print("sf : ",uniform.sf(x, loc=a, scale=(b-a))*100)
print("*"*25)
print("mean: ", (a+b)/2)
print("vari: ", (b-a)**2/12)
print("std : ", np.sqrt((b-a)**2/12))
print("*"*25)
print("los resultados obtenidos son iguales al ejercicio del cuadro anterior")
```

# 2. <font color='green'>Distribución de Bernoulli</font>

1. Modelo para representar una **variable aleatoria discreta**, la cual solo puede tener dos sucesos posibles: exito o fracaso. 

2. Estos sucesos son **mutuamente excluyentes**.

3. Se realiza un solo experimento (n=1)

4. El fracaso (no éxito) ***no es lo contrario de exito, sino a cualquier caso distinto a exito***, siempre y cuando haya más de dos probabilidades.

* Exito (p) = P(X=1)
* Fracaso (q) = P(X=0) = 1 - p

## Ejemplos teóricos:

1. lanzamiento de moneda (se efectúa una vez, 1/2)
2. lancamiento de un datos (se efectúa una vez, 1/6)


## Ejemplo más elaborado:
Carrera de ciclistas: apostamos a un corredor. Pero son 10 corredores

Existe un sólo éxito posible: que gane nuestro corredor. Pero el fracaso (no éxito), en este caso, es más amplio.

**P(X=1) = 1/10**

Es decir: **P = casos probables / casos posibles** (regla de Laplace)

```python
from scipy.stats import bernoulli as berno
import matplotlib.pyplot as plt
import numpy as np
```

## 0.1 ejercicios: ciclista
**manualmente**

```python
k = 1
m = 10
p = (k/m)**k * (m-1/m)**(k-1)
print("exito: ", p)
print("fracaso: ", 1-p)
```

**scipy**

```python
# ingreso
n = 1 #nuestra apuesta
p = 1/10 #probable/posible
loc=0

# PROCEDIMIENTO
k = np.arange(loc-1, n+1)
px = berno.pmf(k, p, loc)

mean, var, skew, kurt = berno.stats(px, moments='mvsk')
print("mean={}\nvar={}\nskew={}\nkurt={}".format(mean, var, skew, kurt))

print("*"*25)
print("pmf :", berno.pmf(k,p,loc))
print("cdf :", berno.cdf(k,p,loc))
print("sf  :", berno.sf(k,p,loc))

# SALIDA
plt.title('bernoulli rango hasta n='+str(n))
plt.stem(k, px, use_line_collection=True)
plt.margins(0.1)
plt.show()
```

# 1. <font color='green'>Distribución Binomial</font>

**Distribución de probabilidad (discreta) que describe el número de `exitos` al realizar `N experimentos independientes entre sí`**

se entiende como una serie de pruebas en la que solo podemos tener dos resultados: `exito o fracaso`, siendo el `exito` la variable aleatoria

**Propiedades:**

1. dos resultados posibles: éxito o fracaso
2. las probabilidades de éxitos y fracasos deben ser constantes
3. cada resultado es independiente del experimento anterior
4. los resultados son mutuamente excluyentes
5. los sucesos son exahustivos: al menos uno de los dos debe ocurrir



1. E[X] = esperanza = media = n*p
2. V[X]varianza = n * p * q = E[X] * q
3. STD = RaizCuadrada V[X]

## 1.1 ejercicio:
Ejemplo: se fabrican un 5% de productos defectuosos. Sobre una muestra de 12 productos ¿qué probabilidad existe que 2 de ellos sean defectuosos?
* x = variable apredecir
* k = 2 (la cantidad de éxitos=obtener productos defectuosos)
* n = 12 (muestra)
* p = 0.05 (5%=promedio del éxito=productos defectuosos)
* q = 0.95 (1-p=probabilidad del fracaso)

### manualmente:
## P(X=2) = (12! 2!) * 0.05^2 * (1 - 0.95)^12-2
## P(X=2) = (12 ! 2!) * 0.0025 * 0.005^10
## P(X=2) = (136/2) * 0.0025 * 0.005^10
## P(X=2) = 0.0988 = 9.88% de probabilidad



### scipy

```python
from scipy.stats import binom as binom
```

```python
k, n, p = 2, 12, 0.05 #exitos esperados, muestra y % de productos defectuosos
bn = binom.pmf(k,n,p)
print("probabilidad: ", np.around(bn*100, 2))
mean, var, skew, kurt = binom.stats(k, p, moments='mvsk')
print("mean={}\nvar={}\nskew={}\nkurt={}".format(mean, var, skew, kurt))

x = np.arange(0,13,1)
pmf = binom.pmf(x,n,p)
plt.bar(x,pmf, color='orange')
plt.xlabel('muestras')
plt.ylabel('probabilidad')
plt.title('gráfica')
```
