```python
import scipy.stats as ss
import numpy as np
```

**Ejercicios simples:**

calcular la probabilidad de que **x** se encuentre por debajo de un valor dado

```python
#variable, media y std
x, m, s = 8, 10, 2 
normal = ss.norm(m, s)
pdf = normal.pdf(x)
cdf = normal.cdf(x)
print("pdf: ", np.around(pdf,3))
print("cdf: ", np.around(cdf,3))
```

La temperatura durante setiembre est√° distribuida normalmente con media 18,7¬∫C
y desviaci√≥n standard 5¬∫C. Calcule la probabilidad de que la temperatura durante setiembre est√©
por debajo de 21¬∫C. 

```python
x, m, s = 21, 18.7, 5 
normal = ss.norm(m, s)
pdf = normal.pdf(x)
cdf = normal.cdf(x)
print("pdf: ", np.around(pdf,3))
print("cdf: ", np.around(cdf,3))
```

los resultados de un examen siguen una distribuci√≥n normal con media 78 y desviaci√≥n t√≠pica 36
probabilidad de que una persona obtenga una calificaci√≥n superior a 72

```python
x, m, s = 72, 78, 36 
normal = ss.norm(m, s)
pdf = normal.pdf(x)
cdf = normal.cdf(x)
print("pdf: ", np.around(1-pdf,3))
print("cdf: ", np.around(1-cdf,3))
```

test de IQ: media=100, std=15

que x >= 95 x <= 110

```python
x, m, s = 110, 100, 15  #primero calculo los IQ < 110
normal = ss.norm(m, s)
cdf1 = normal.cdf(x)

x = 95
normal = ss.norm(m, s)
cdf2 = 1 - normal.cdf(x) #con 1-cdf calculo los IQ > 95

print(np.around(cdf1,2))
print(np.around(cdf2,2))

res = cdf1 - (1-cdf2) #obtengo el total
print("resultado: ", np.around(res, 3))
```
