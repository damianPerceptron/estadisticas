# Estadistica

### Intro:

Emplearemos Medidas de Tendencia Central y Medidas de Variabilidad.

La función de las Medidas de Tendencia Central es proveer informacion descriptiva sobre el valor numerico que es considerado el mas usual para una variable cuantitativa.

Las Medidas de tendencia central son:

* **Media**

Dados los n numeros ${\{x_{1},x_{2},\ldots ,x_{n}\}}$ la media aritmetica se define como:

\begin{equation}
    \bar {x} = {\frac {1}{n}} \sum _{i=1}^{n}x_{i} ={\frac {x_{1}+x_{2}+\cdots +x_{n}}{n}}
\end{equation}

* **Mediana**

Sean ${ x_{1},x_{2},x_{3},\ldots ,x_{n} }$ los datos de una muestra ordenada en orden creciente y designando la mediana como $M_{e}$, distinguimos dos casos:

a) Si n es impar, la mediana es el valor que ocupa la posición $(n+1)/2$ una vez que los datos han sido ordenados (en orden creciente o decreciente), porque este es el valor central. Es decir: $M_{e}=x_{{(n+1)/2}}$.

b) Si n es par, la mediana es la media aritmética de los dos valores centrales. Cuando $n$ es par, los dos datos que están en el centro de la muestra ocupan las posiciones $n/2$ y $(n/2)+1$. Es decir: $M_{e}=(x_{{{\frac{n}{2}}}}+x_{{{{\frac{n}{2}}}+1}})/2$.

* **Moda**

Es el valor del dato con mayor frecuencia en un conjunto.

**Medidas de Variabilidad:**

Las medidas de variabilidad o dispersión nos indican si los valores de los datos están próximos entre sí o si 
por el contrario están muy dispersos. Estas medidas se determinan en función de la distancia entre los datos y algún estadístico de tendencia central.

* **Rango**

Sean ${ x_{1},x_{2},x_{3},\ldots ,x_{n} }$ los datos de una muestra ordenada en orden creciente, el rango es $x_{n} - x_{1}$

* **Varianza**

Si tenemos un conjunto de valores de una variable, la varianza se calcula de la siguiente forma:

\begin{equation}
\sigma_{n}^{2} = \frac{1}{n} \sum _{i=1}^{n} (x_i - \bar{X})^{2}
\end{equation}

Siendo:

$x_{i}$: cada dato

$\bar{X}$: media de los datos

$n$: número de datos

* **Desvio Estandar**

Es la raíz cuadrada de la varianza

\begin{equation}
\sigma_{n} = \sqrt{\frac{1}{n} \sum _{i=1}^{n} (x_i - \bar{X})^{2}}
\end{equation}

* **Coeficiente de Variacion**

Es el desvío estándar dividido por la media

\begin{equation}
CV = \frac{\sigma_{n}}{\bar{X}} . 100 
\end{equation}

**NumPy tiene funciones para calcular todas estas medidas**

Ejercicio 1: vamos a calcular la media de cada una de las poblaciones usando la matriz `poblaciones` y el metodo `mean` de NumPy rendondeando a dos decimales.

Pregunta: Que valor de `axis` debemos usar? porque?

```python
print ("Conejos, Linces, Zanahorias")
print ("Mean:", np.around(poblaciones.mean(axis=0), decimals=2))
print ("Std:", np.around(poblaciones.std(axis=0), decimals=2))
```

*Respuesta: Cada poblacion esta representada por una columna en la matriz poblaciones, por lo tanto para calcular la media por poblacion debemos colapsar las filas, y para eso el valor de axis debe ser 0*

A continuacion, calculamos para cada especie el año en el que tuvieron maxima poblacion.

Utilizamos el metodo `argmax` de NumPy que devuelve el indice en el array del valor maximo.

Igual que en el punto anterior, las poblaciones están representadas por columnas, entonces debemos colapsar filas y así axis debe tener valor 0. 

`indice_max_poblacion` tendrá tantos elementos como poblaciones considero (3 en este caso: conejos, linces, zanahorias) y cada elemento será **el índice de la fila** donde esa población tuvo la máxima cantidad de individuos.

```python
indice_max_poblacion = np.argmax(poblaciones, axis=0)
indice_max_poblacion
```

Por lo tanto, la población de conejos fue máxima en la cuarta fila, la de linces en la quinta fila, la de zanahorias en la primer fila.

Veamos a que años corresponden esas filas, usando Fancy indexing

https://docs.scipy.org/doc/numpy/reference/generated/numpy.argmax.html

```python
annos_con_maximos = anno[indice_max_poblacion]
annos_con_maximos
```

*Entonces la máxima población de conejos fue en 1903, la de linces fue en 1904, y la de zanahorias fue en 1900*

```python
# Fancy Indexing
print ("Conejos, Linces, Zanahorias")
print ("Años de máxima población:", anno[indice_max_poblacion])
```

### COVARIANZA:

La covarianza es un valor que indica el grado de variacion conjunta de dos variables aleatorias respecto a sus medias. 

Es el dato basico para determinar si existe una dependencia entre ambas variables.

El metodo que calcula la covarianza en NumPy es `cov`. 

Bibliografia oficial: https://docs.scipy.org/doc/numpy/reference/generated/numpy.cov.html

```python
print(np.around(np.cov([conejos, linces, zanahorias]), decimals=2))
print("*"*50)
print("conejos: ", 4.58558619e+08)
print("linces: ",  2.77422333e+08)
print("zanahorias: ", 1.15910000e+07)
```

### Correlacion:

El coeficiente de correlacion de Pearson es una medida lineal entre dos variables aleatorias cuantitativas.

A diferencia de la covarianza, la correlacion de Pearson es independiente de la escala de medida de las variables.

Podemos definir el coeficiente de correlacion de Pearson como un indica que puede utilizarse para medir el grado de relacion de dos variables siempre y cuando ambas sean cuantitativas.

El metodo que calcula la correlacion de Pearson en NumPy es `corrcoef`. Los valores de los elementos de la matriz de correlacion estan entre -1 y 1 inclusive. Es equivalente a la matriz de covarianzas normalizada.

Bibliografia oficial: https://docs.scipy.org/doc/numpy/reference/generated/numpy.corrcoef.html

```python
print(np.corrcoef([conejos, linces, zanahorias]))
print("*"*50)
print(np.cov([conejos, linces, zanahorias]))
```

```python
from sklearn.linear_model import LinearRegression
def graficarRegresion(x, y, l1, l2):
    lr = LinearRegression()
    lr.fit(x, y)
    b0=lr.intercept_
    b1=lr.coef_[0]
    print("b0 sklearn: ", b0,"\nb1 sklearn: ", b1)
    y_pred_m = b0 + b1 * x
    print("predicción manual: ", np.around(np.mean(y_pred_m),2))
    y_pred_s = lr.predict(x)
    print("predicción sklearn: ", np.around(np.mean(y_pred_s),2))
    plt.scatter(x, y, color = 'red')
    plt.plot(x, y_pred_m, color = 'blue')
    plt.title(l1 + "/" + l2)
    plt.ylabel(l1)
    plt.xlabel(l2)
    plt.show()    
    
    
```
