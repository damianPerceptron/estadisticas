# Distribución <font color='red'>Binomial</font>

**Distribución de probabilidad (discreta) que describe el número de `exitos` al realizar `N experimentos independientes entre sí`**

se entiende como una serie de pruebas en la que solo podemos tener dos resultados: `exito o fracaso`, siendo el `exito` la variable aleatoria

**Propiedades:**

1. dos resultados posibles: éxito o fracaso
2. las probabilidades de éxitos y fracasos deben ser constantes
3. cada resultado es independiente del experimento anterior
4. los resultados son mutuamente excluyentes
5. los sucesos son exahustivos: al menos uno de los dos debe ocurrir

## fórmula:

## X~Bi(10, 0.1) = distr. binomial con 10 experimentos y 10% de probabilidad

1. E[X] = esperanza = media = n*p
2. V[X]varianza = n * p * q = E[X] * q
3. STD = RaizCuadrada V[X]

* x = variable a predecir
* k = (o `r`) el valor que toma la variable
* n = número de veces que se repetirá el experimento
* p = la probabilidad del éxito (probabilidad que se obtenga `x`)
* q = probabilidad del fracaso

Ejemplo: se fabrican un 5% de productos defectuosos. Sobre una muestra de 12 productos ¿qué probabilidad existe que 2 de ellos sean defectuosos?
* x = variable apredecir
* k = 2 (la cantidad de éxitos=obtener productos defectuosos)
* n = 12 (muestra)
* p = 0.05 (5%=probabilidad del éxito=productos defectuosos)
* q = 0.95 (1-p=probabilidad del fracaso

## P(X=2) = (12! 2!) * 0.05^2 * (1 - 0.95)^12-2

## P(X=2) = (12 ! 2!) * 0.0025 * 0.005^10

## P(X=2) = (136/2) * 0.0025 * 0.005^10

## P(X=2) = 0.0988 = 9.88% de probabilidad

```python
import scipy.stats as ss
import numpy as np
import matplotlib.pyplot as plt
```

## ejemplo: productos defectuosos; muestra de 12, probabilidad 0.05

```python
k, n, p = 2, 12, 0.05
binomial2 = ss.binom.pmf(k, n, p) * 100
print(np.around(binomial2,2))

mean, var, skew, kurt = ss.binom.stats(12, p, moments='mvsk')
print("mean ", mean)
print("var  ", var)
print("skew ", skew)
print("kurt ", kurt)
```

```python
x = np.arange(0,11,1)
pmf = ss.binom.pmf(x,12,0.05)
plt.bar(x,pmf, color='orange')
plt.xlabel('muestras')
plt.ylabel('probabilidad')
plt.title('gráfica')
```

```python
n = ss.binom(12, 0.05) #12=muestra; 0.05=probabilidad de éxito (productos defectuosos)
k = 2 #2=probabilidad de obtener 2 éxitos en la muestra (12)
binomial = n.pmf(k)
print(np.around(binomial*100,2))
print(binomial, 1-binomial)
```

## Un jugador marca el 85% de los penaltis que intenta. Si lanza 8 penaltis calcular la probabilidad de

a) Marque más de 6 penaltis

b) Marque al menos 6 penaltis

```python
#marque más menos 6
binomial2=0
n=8
p=0.85
for k in range(7,9):
    binomial2 += ss.binom.pmf(k, n, p) * 100
    
print(np.around(binomial2,2))

mean, var, skew, kurt = ss.binom.stats(n, p, moments='mvsk')
print("mean ", mean)
print("var  ", var)
print("skew ", skew)
print("kurt ", kurt)
```

```python
x = np.arange(0,11,1)
pmf = ss.binom.pmf(x,8,0.85)
plt.bar(x,pmf, color='orange')
plt.xlabel('muestras')
plt.ylabel('probabilidad')
plt.title('gráfica')
```

```python
#que marque al menos 6
binomial2=0
n=8
p=0.85
for k in range(6,9):
    binomial2 += ss.binom.pmf(k, n, p) * 100
    
print(np.around(binomial2,2))

mean, var, skew, kurt = ss.binom.stats(n, p, moments='mvsk')
print("mean ", mean)
print("var  ", var)
print("skew ", skew)
print("kurt ", kurt)
```

## ejemplo: El 5% de los clientes de una entidad bancaria son morosos. ¿Cuál es la probabilidad de encontrar al menos un moroso entre 10 clientes elegidos al azar?
