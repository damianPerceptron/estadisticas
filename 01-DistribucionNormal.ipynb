#### Paso 2: Parametros

```python
# desviación estándar muestral y media
std = np.std(data('cancer')['age'], ddof = 1)
mean = np.mean(data('cancer')['age'])

print("mean ", mean, "std ", std)
```

```python
domain = np.linspace(np.min(data('cancer')['age']), np.max(data('cancer')['age']))
#plt.plot(domain, norm.pdf(domain, mean, std),
#        label = '$\mathcal{N}$ ' + f'$( \mu \\approx {round(mean)}, \sigma \\approx {round(std)} )$')

plt.plot(domain, norm.pdf(domain, mean, std), label="hola")

plt.hist(data('cancer')['age'], edgecolor = 'black', alpha = .5, density = True)
plt.title('Normal Fit')
plt.xlabel('Valor')
plt.ylabel('Densidad')
#plt.legend()
plt.show()
```

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu = 62.44736 #mean
sigma = 9.07345 #std
x = 70 #edada a predecir
y = 0

#CON LA RESTA CALCULO LA PROBABILIDAD ENTRE UN RANGO:
cdf_probability = norm.cdf(x, mu, sigma) - norm.cdf(y, mu, sigma)
#PERO ESTE RESULTADO ME ARROJA LA PROBABILIDAD "A LA IZQUIERDA", O SEA <= AL SUELDO
print("probabilidad 1: ", np.around(cdf_probability*100,3), " (edad <=",x,")")

#OTRA FORMA, SIN EL -1
print("probabilidad 2:", np.around(norm.sf(x, mu, sigma)*100,3)," (edad >=",x,")")

mean, var, skew, kurt = norm.stats(moments='mvsk')
print("mean: ", mean)
print("variance: ", var)
print("skew: ", skew)
print("kurtosis: ", kurt)
```
